```
import pandas as pd

x0 = ['A','B','C','D','E', 'F', 'G']
y0 = [10, 20, 30, 40, 5, 6, 7]

df0 = pd.DataFrame( data=list(zip(x0,y0)), columns=['produto_loja_A', 'Preco'] )

x1 = ['D','E','F','G','H','I','J', 'H', 'I']
y1 = [30, 40 , 1, 3, 20, 4, 5, 10, 11]

df1 = pd.DataFrame( data=list(zip(x1,y1)), columns=['produto_loja_B', 'Preco'] )
```

# Inner Join

```
# Left Table - esquerdo
# Right Table - direito
df_resultante = pd.merge(left = df0, right = df1, how='inner', left_on='produto_loja_A', right_on='produto_loja_B')
df_resultante
```

# Outer Join

```
# Left Table - esquerdo
# Right Table - direito
df_resultante = pd.merge(left = df0, right = df1, how='outer', left_on='produto_loja_A', right_on='produto_loja_B')
df_resultante
```

# Left Join

```
# Left Table - esquerdo
# Right Table - direito
df_resultante = pd.merge(left = df0, right = df1, how='left', left_on='produto_loja_A', right_on='produto_loja_B')
df_resultante
```

# Right Join

```
# Left Table - esquerdo
# Right Table - direito
df_resultante = pd.merge(left = df0, right = df1, how='right', left_on='produto_loja_A', right_on='produto_loja_B')
df_resultante
```

# Dataframe.merge

```
df0.merge(df1, how='inner', left_on='produto_loja_A', right_on='produto_loja_B')
```

```
#Atente-se apenas ao comando
df0.merge(df1, how='inner', on='Preco')
```
